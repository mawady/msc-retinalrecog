# Dataset Samples

```python
import tensorflow as tf
from tensorflow import keras
import matplotlib as mp
from matplotlib import pyplot as plt 
import numpy as np
import random
import cv2
import os

plt.figure(figsize=(20,20))

print("Kaggle APTOS2019:")
aptos2019_dir='./Datasets/APTOS2019/train_images'

for i in range(5):
    file = random.choice(os.listdir(aptos2019_dir))
    image_path= os.path.join(aptos2019_dir, file)
    img=mp.image.imread(image_path)
    ax=plt.subplot(1,5,i+1)
    ax.title.set_text(file)
    plt.imshow(img)
```
